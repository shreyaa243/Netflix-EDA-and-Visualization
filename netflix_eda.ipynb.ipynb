## DATA CLEANING STEPS

```python
import pandas as pd
```

```python
#remove or fill values
df.fillna("Unknown", inplace=True)
```

```python
df
```

```python
# Data types
df.dtypes
```

```python
# Drop duplicates
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
# Strip any spaces from column names
df.columns = df.columns.str.strip()
```

```python
#make all string lowercase
df['Title'] = df['Title'].str.lower()
```

```python
#convert and extract year from release date
df['Release_Date'] = pd.to_datetime(df['Release_Date'], errors='coerce')
df['Year'] = df['Release_Date'].dt.year
```

```python
# Null values
"Null values:"
df.isnull().sum()
```

## Time based analysis

```python
#extract
df['Year'] = df['Release_Date'].dt.year
df['Month'] = df['Release_Date'].dt.month
df['Day'] = df['Release_Date'].dt.day
```

```python
#total releases by year
df['Year'].value_counts().sort_index()
```

```python
#breakdown by type over the years
df.groupby(['Year', 'Type']).size()
```

```python
#what month see the most releases
df['Month'].value_counts().sort_index()
```

```python
#most common release day
df['Release_DayName'] = df['Release_Date'].dt.day_name()
df['Release_DayName'].value_counts()
```

```python
#recent years focus
df[df['Year'] >= 2018]['Year'].value_counts().sort_index()
```

```python
#shows full time span of netflix data
df['Release_Date'].min(), df['Release_Date'].max()
```
